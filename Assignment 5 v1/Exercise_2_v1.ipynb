```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n_steps = 300
x = np.zeros(n_steps + 1)  
x[0] = 5
for k in range(1, n_steps):
    x[k] = 0.9 * x[k - 1] + np.random.normal(0, np.sqrt(0.19) )

plt.figure(figsize=(10, 6))
plt.hist(x, bins=15, density=True, alpha=0.6, label="Simulated Histogram")

x_vals = np.linspace(-4, 4, 500)
plt.plot(x_vals, norm.pdf(x_vals, loc=0, scale=1), label="Standard Normal PDF", linewidth=2)

plt.title("Histogram of Simulated Markov Chain vs Standard Normal PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(x, label='Markov chain trajectory')
plt.axhline(0, color='r', linestyle='--', label='Mean of standard normal distribution.')
plt.xlabel('Step')
plt.ylabel('x')
plt.title('Markov Chain Trajectory')
plt.legend()
plt.show()
```
